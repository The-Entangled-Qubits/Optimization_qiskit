# Teleportation

This section talks about quantum teleportation. WE will be building a quantum circuit to implement teleportaion using basic quantum gates. This process helps in transfering information from one section to another. We cannot copy and paste like we do in classical computers, as copying means measuring a quantity whch leads to loss of data from that source in quantum computing. We use quantum entanglement here to complete the process.

```python
from qiskit import *
```

```python
cir = QuantumCircuit(3,3) #3 qubits and 3 classical bits
```

```python
%matplotlib inline
cir.draw(output = 'mpl')
```

```python
cir.x(0)
cir.barrier()
cir.draw(output = 'mpl')
```

```python
#creating entaglement between q1 and q2
cir.h(1)
cir.cx(1,2)
cir.draw(output = 'mpl')
```

```python
cir.cx(0,1)
cir.h(0)
cir.draw(output = 'mpl')
```

```python
#Measurement on q0 and q1
cir.barrier()
cir.measure([0,1], [0,1])
cir.draw(output = 'mpl')
```

```python
cir.cx(1,2)
cir.cz(0,2)
cir.draw(output = 'mpl')
```

The state q0 has been teleported to q2. Measurement on q2 can be used to prove this. The expected result is 1 in q2. 

```python
sim = Aer.get_backend('qasm_simulator')
```

```python
cir.measure(2,2)
m= execute(cir, backend = sim, shots = 1024). result()
number = m.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(number)
```

The results have been expressed in the form of q2,q1,q0. As we see, the value of q2 is always 1 regardless of the values of q0 and q2, hence we can say that the information has been transferred accurately.
